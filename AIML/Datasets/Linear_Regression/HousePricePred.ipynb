```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\HP\\Desktop\\AIML\\Datasets\\homeprices.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.Area, df.Price, color='Orange', marker='*')
plt.show()
```

```python
new_df = df.drop('Price', axis='columns')
new_df
```

```python
price = df.Price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

# (1) Prdict price of a home with area = 3300 sqr ft

```python
reg.predict([[3300]])
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
